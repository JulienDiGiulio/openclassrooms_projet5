## <u>Reflexion suite à l'analyse</u>

Olist souhaite que vous fournissiez à ses équipes d'e-commerce une segmentation des clients qu’elles pourront utiliser au quotidien pour leurs <u>campagnes de communication</u>.

Votre objectif est de comprendre les <u>différents types d’utilisateurs</u> grâce à <u>leur comportement et à leurs données personnelles</u>.

<u>Prenons connaissance des différentes méthodes de segmentations des clients :</u>

![title](https://res.cloudinary.com/dyd911kmh/image/upload/f_auto,q_auto:best/v1537803937/types_of_segmentation_hm7mmz.png)

 - Nous n'avons aucune donnée disponible pour de la segmentation démographique.
 - Nous pouvons faire de la segmentation géographique (Localisation, ville, état).
 - Nous pouvons faire de la segmentation comportementale (Dépenses, type de produits, catégories de produits).
 - Nous pouvons faire de la segmentation psychologique (commenataires).
 
<u>Aspects techniques</u>

Le nombre de clients sur le site est identifié par la variable customer_unique_id.

La variable customer_id représente un numéro d'identifiant par facture.

Pour une étude classification notre base de données doit être construite en conséquence :
 - 1 ligne = 1 cluster. Chaque ligne représente donc 1 client du site.
 - Toutes les informations recherchées sur chaque client doivent donc être ramenées à une valeur numérique par feature.

## <u>Pipeline de construction des features</u>

### Données de géolocalisation

```python
def features_geolocation(data):
    
    #On récupère ensuite les données géographiques présentes dans les données d'entrées
    df_zip = data_geolocation.loc[:,["geolocation_zip_code_prefix", "geolocation_lat", "geolocation_lng"]]
    print("Il y a", len(df_zip["geolocation_zip_code_prefix"].unique()), "codes postaux renseignés.")

    #On calcule le baricentre des coordonnées par zip code
    df_zip = df_zip.groupby(["geolocation_zip_code_prefix"], as_index=False).mean()
    
    #On renomme les clés à l'identique pour la jointure
    data.rename(columns={"customer_zip_code_prefix": "zip_code_prefix"}, inplace=True)
    df_zip.rename(columns={"geolocation_zip_code_prefix": "zip_code_prefix"}, inplace=True)
    #print(data["zip_code_prefix"].isin(df_zip["zip_code_prefix"]).value_counts())

    #On joint les 2 dataframes
    data = pd.merge(data, df_zip, on=["zip_code_prefix"], how="inner")
    
    return data
```

```python
Categ_regions = {
    "DF": "centre_ouest",
    "GO": "centre_ouest",
    "MS": "centre_ouest",
    "MT": "centre_ouest",
    
    "MA": "nord_est",
    "PI": "nord_est",
    "RN": "nord_est",
    "PB": "nord_est",
    "CE": "nord_est",
    "AL": "nord_est",
    "PE": "nord_est",
    "SE": "nord_est",
    "BA": "nord_est",
    
    "AC": "nord",
    "AM": "nord",
    "RO": "nord",
    "RR": "nord",
    "PA": "nord",
    "AP": "nord",
    "TO": "nord",
    
    "MG": "sud_est",
    "ES": "sud_est",
    "RJ": "sud_est",
    "SP": "sud_est",
    
    "PR": "sud",
    "SC": "sud",
    "RS": "sud"
}
   
lst_regions = [
    "centre_ouest",
    "nord_est",
    "nord",
    "sud_est",
    "sud"
]
```

```python
def regions(data):
    
    df_regions_categ = data_customer.loc[:,["customer_zip_code_prefix", "customer_state"]]

    df_regions_categ["customer_state"].replace(Categ_regions, inplace=True)
    
    #On joint les 2 dataframes
    data = pd.merge(data, df_regions_categ, on=["customer_zip_code_prefix"], how="inner")
    
    data.rename(columns={"customer_state": "region_client"}, inplace=True)
    return data
    
```

### Prix commandes + prix FDP

Cette fonction permet de calculer le montant total dépensé par client par facture.
Idem pour le total des frais de port.

```python
def features_prices(data):

    df_prices = data_order_items.loc[:, ["order_id", "price", "freight_value"]]
    
    print("Il y a", len(df_prices["order_id"].unique()), "commandes renseignés.")

    #On calcule la moyenne des dépenses
    df_prices = df_prices.groupby(["order_id"], as_index=False).mean()
    
    #On joint les 2 dataframes
    data = pd.merge(data, df_prices, on=["order_id"], how="inner")
    
    return data
```

```python
def features_product_id(data):

    df_product_id = data_order_items.loc[:, ["order_id", "product_id"]]
    
    data = pd.merge(data, df_product_id, on=["order_id"], how="inner")
    
    return data
```

### Crédits + Type de paiement

Cette fonction permet de calculer le type de paiement le plus fréquent utilisé.
Elle crée aussi une feature booléenne afin d'identifier si le client à payer à crédit ou non.

```python
def feature_credit(data):
    
    df_credit = data_order_payments.loc[:,["order_id", "payment_installments"]]
    
    #On transforme la variable de paiements étalés en booléen
    #df_credit["payment_installments"]=df_credit["payment_installments"].replace(to_replace=range(1,29), value=1)
    #df_credit["payment_installments"].astype("bool")
    
    #On joint les 2 dataframes
    data = pd.merge(data, df_credit, on=["order_id"], how="inner")
        
    return data
```

```python
def features_payments(data):
    
    df_payments = data_order_payments.loc[:,["order_id", "payment_type"]]
    
    #on récupère le type de paiement le plus fréquent
    #df_payments = df_payments.groupby(["order_id"], as_index=False).value_counts().idxmax()
    df_payments = df_payments.groupby(["order_id"], as_index=False)["payment_type"].agg(pd.Series.mode)
    
    
    #On joint les 2 dataframes
    data = pd.merge(data, df_payments, on=["order_id"], how="inner")
    
    
    
    return data
    
```

```python
Categ_payment = {
    "credit_card": "Carte_bancaire",
    "debit_card": "Carte_bancaire", 
    "'credit_card' 'debit_card'": "Carte_bancaire",
    
    "boleto": "liquide",
    "'credit_card' 'voucher'": "Carte_bancaire_bons_reductions",
    "voucher": "bons_reductions"
}
    
lst_payment = ["Carte_bancaire",
                "liquide",
                "Carte_bancaire_bons_reductions",
                "bons_reductions"
                ]
```

```python
#Cette fonction permet de catégoriser les tyes paiements avec de nouvelles valeurs

def feature_category_payment(data):
    
    
    df_payment_categ = data_order_payments.loc[:,["order_id", "payment_type"]]
    
    #Mais avant nous devons traiter la colonne avec les types de paiement car certaines valeurs contiennent des crochets
    #et font bugguer la fonction
    data["payment_type"] = data["payment_type"].astype(str)
    data["payment_type"] = data["payment_type"].str.strip("[]")
    
    data.drop("payment_type", axis=1, inplace=True)
    df_payment_categ["payment_type"].replace(Categ_payment, inplace=True)
    
    #On joint les 2 dataframes
    data = pd.merge(data, df_payment_categ, on=["order_id"], how="inner")
    
    return data
```

### Commentaires

Cette fonction crée une feature booléenne afin de savoir si le client a déjà laissé un commentaire ou non sur le site.

```python
def feature_comment(data):
    
    df_comment = data_order_reviews.loc[:,["order_id", "review_comment_message"]]
    
    #On transforme la variable de paiements étalés en booléen
    df_comment["review_comment_message"] = df_comment["review_comment_message"].apply(lambda x: 1 if not pd.isnull(x) else 0)
    df_comment["review_comment_message"].astype("bool")
    
    #On joint les 2 dataframes
    data = pd.merge(data, df_comment, on=["order_id"], how="inner")
        
    return data
```

### Encodage OneHot des catégories

Cette fonction crée des features binaires qui nous permettrons de visualiser quels produits le client a déjà acheté sur le site.

```python
def Encodage(data):
    
    print("Column headings:")
    print(data.columns)

    # Categorical boolean mask
    categorical_feature = ["payment_type", "product_category_name_english"]
    
    # filter categorical columns using mask and turn it into a list
    #categorical_cols = data.columns[categorical_feature].tolist()
    
    print(data.shape)
    
    data[categorical_feature] = data[categorical_feature].astype("category")
    
    data2 = pd.get_dummies(data[categorical_feature])
    
    #On joint les 2 dataframes
    data = pd.concat([data, data2], axis=1)
    
    return data
```

### Nombre de produits

Cette fonction permet de créer des features recensant chaque catégorie de produits.

Elles nous seront utiles pour l'aggrégation des données par client.

```python
def create_features_categories(data):
    
    #On crée les colonnes correspondantes aux catégories de type de paiement
    for x in lst_regions:
        
        col = "region_" + x
        data[col] = np.where(data["region_client"] == x, 1, 0)
        
    
    #On crée les colonnes correspondantes aux catégories de type de paiement
    for x in lst_payment:
        
        col = "taux_paiement_" + x
        data[col] = np.where(data["payment_type"] == x, 1, 0)
        
        col = "nbr_paiements_effectues"
        data[col] = 1
    
    
    #On crée les colonnes correspondantes aux catégories de produits
    for x in lst_categ:
        
        col = "taux_produit_" + x
        data[col] = np.where(data["product_category_name_english"] == x, 1, 0)
        
        col = "nbr_produits_achetes"
        data[col] = 1
    
    return data
    
```

## <u>Pipeline de construction des features</u>

### Nombre de commandes

```python
def nombre_commandes(data_prepared):

    #Nombre de commandes
    #nbr_commandes = data.loc[:,["customer_unique_id", "order_id"]]
    
    nbr_commandes = pd.DataFrame(data.groupby(["customer_unique_id"])["order_id"].nunique())
    
    nbr_commandes.rename(columns={"index": "customer_unique_id", "order_id": "nbr_commandes"}, inplace=True)
    
    #On crée des catégories représentant des tranches de commandes passées sur le site
    nbr_commandes["juste_1_commandes"] = np.where(nbr_commandes["nbr_commandes"] == 1,
                                                 1, 0)
    
    nbr_commandes["de_2_4_commandes"] = np.where((nbr_commandes["nbr_commandes"] >= 2) &
                                                 (nbr_commandes["nbr_commandes"] < 4),
                                                 1, 0)
    
    nbr_commandes["de_4_6_commandes"] = np.where((nbr_commandes["nbr_commandes"] >= 4) &
                                                 (nbr_commandes["nbr_commandes"] < 6),
                                                 1, 0)
    
    nbr_commandes["de_6_8_commandes"] = np.where((nbr_commandes["nbr_commandes"] >= 6) &
                                                 (nbr_commandes["nbr_commandes"] < 8),
                                                 1, 0)
    
    nbr_commandes["de_8_10_commandes"] = np.where((nbr_commandes["nbr_commandes"] >= 8) &
                                                 (nbr_commandes["nbr_commandes"] < 10),
                                                 1, 0)
    
    nbr_commandes["de_10_commandes_et_plus"] = np.where((nbr_commandes["nbr_commandes"] > 10),
                                                 1, 0)
    
    #On joint les 2 dataframes
    data_prepared = pd.merge(data_prepared, nbr_commandes, on=["customer_unique_id"], how="inner")
    
    return data_prepared
```

### Data d'achat la plus récente

Cette fonction permet de récupérer la dernière commande qu'a effectué le client sur le site

```python
def date_achat_recent(data_prepared):
    
    #On récupère les données de notre table unique
    df_date_achat = data.loc[:,["customer_unique_id", "order_purchase_timestamp"]]
    
    #On convertit le champs en date
    df_date_achat["order_purchase_timestamp"] = pd.to_datetime(df_date_achat["order_purchase_timestamp"])
    df_date_achat["order_purchase_timestamp"] = pd.to_datetime(df_date_achat["order_purchase_timestamp"]).dt.strftime('%d%m%Y').astype("int64")
    
    #On recherche la date la plus récente par customer_unique_id
    df_date_achat = df_date_achat.groupby(["customer_unique_id"], as_index=False)["order_purchase_timestamp"].min()
    
    df_date_achat.rename(columns={"order_purchase_timestamp": "date_derniere_commande"}, inplace=True)
        
    #On joint les 2 dataframes
    data_prepared = pd.merge(data_prepared, df_date_achat, on=["customer_unique_id"], how="inner")
    
    return data_prepared
```

### Ancienneté client

Dans cette fonction, nous allons faire la différence entre la date d'aujourd'hui et la date de première commande de chaque client en mois.

```python
from datetime import date

def anciennete_client(data_prepared):
    
    #On récupère les données de notre table unique
    df_anciennete = data.loc[:,["customer_unique_id", "order_purchase_timestamp"]]
    
    #On va prendre comme date de fin la commande la plus récente du dataset
    date_plus_recente = pd.to_datetime(df_anciennete["order_purchase_timestamp"]).dt.strftime('%d/%m/%Y')
    date_plus_recente = pd.to_datetime(date_plus_recente)
    date_plus_recente = date_plus_recente.max()
    
    #On convertit le champs en date
    df_anciennete["order_purchase_timestamp"] = pd.to_datetime(df_anciennete["order_purchase_timestamp"]).dt.strftime('%d/%m/%Y')
    df_anciennete["order_purchase_timestamp"] = pd.to_datetime(df_anciennete["order_purchase_timestamp"])
    
    #On recherche la date la premiere commande de chaque client
    df_anciennete = df_anciennete.groupby(["customer_unique_id"], as_index=False)["order_purchase_timestamp"].max()
    
    #On calcule l'ancienneté du client en jours
    df_anciennete["order_purchase_timestamp"] = pd.to_datetime(df_anciennete["order_purchase_timestamp"])
    df_anciennete["anciennete_client"] = (date_plus_recente - df_anciennete["order_purchase_timestamp"]).dt.days
    
    #df_anciennete.rename(columns={"order_purchase_timestamp": "date_premiere_commande"}, inplace=True)
    df_anciennete.drop("order_purchase_timestamp", axis=1, inplace=True)
    
    #On joint les 2 dataframes
    data_prepared = pd.merge(data_prepared, df_anciennete, on=["customer_unique_id"], how="inner")
    
    return data_prepared
```

### Géolocalisation

Latitude, longitude et régions

```python
def geolocalisation(data_prepared):
    
    #On commence par la latitude et la lingitude
    df_geoloc = data.loc[:,["customer_unique_id", "geolocation_lat", "geolocation_lng"]]
    
    df_geoloc = df_geoloc.groupby(["customer_unique_id"], as_index=False).mean()    
    
    #On joint les 2 dataframes
    data_prepared = pd.merge(data_prepared, df_geoloc, on=["customer_unique_id"], how="inner")
    
    #Ensuite les régions
    
    lst_col = [
        "region_centre_ouest",
        "region_nord_est",
        "region_nord",
        "region_sud_est",
        "region_sud"
    ]
    
    df_geoloc = data[lst_col].copy()
    df_geoloc["customer_unique_id"] = data["customer_unique_id"]
    
    df_geoloc = df_geoloc.groupby(["customer_unique_id"], as_index=False).sum()   
    
    #On joint les 2 dataframes
    data_prepared = pd.merge(data_prepared, df_geoloc, on=["customer_unique_id"], how="inner")
    
    return data_prepared
```

### Dépenses

```python
def depenses(data_prepared):
    
    df_prix = data.loc[:,["customer_unique_id", "price", "freight_value"]]
    
    #dépenses totales sur le site FDP inclus
    df_total = df_prix.groupby(["customer_unique_id"], as_index=False).sum()
    df_total["depenses_totales"] = df_total["price"] + df_total["freight_value"]
    df_total.drop(["price", "freight_value"], axis=1, inplace=True)
    data_prepared = pd.merge(data_prepared, df_total, on=["customer_unique_id"], how="inner")
    
    
    #prix moyen commande FDP inclus
    df_prix_moy = df_prix.groupby(["customer_unique_id"], as_index=False).mean()
    df_prix_moy["prix_moy_commande_fdpin"] = df_prix_moy["price"] + df_prix_moy["freight_value"]
    df_prix_moy["ecart_type_commande_moyenne"] = df_prix_moy["prix_moy_commande_fdpin"].std()
    df_prix_moy.drop(["price", "freight_value"], axis=1, inplace=True)
    data_prepared = pd.merge(data_prepared, df_prix_moy, on=["customer_unique_id"], how="inner")
       
    #prix max d'une commande FDP inclus
    df_commande_max = df_prix.groupby(["customer_unique_id"], as_index=False).max()
    df_commande_max["prix_max_commande_fdpin"] = df_commande_max["price"] + df_commande_max["freight_value"]
    df_commande_max.drop(["price", "freight_value"], axis=1, inplace=True)    
    data_prepared = pd.merge(data_prepared, df_commande_max, on=["customer_unique_id"], how="inner")
    
    #prix min d'une commande FDP inclus
    df_commande_min = df_prix.groupby(["customer_unique_id"], as_index=False).min()
    df_commande_min["prix_min_commande_fdpin"] = df_commande_min["price"] + df_commande_min["freight_value"]
    df_commande_min.drop(["price", "freight_value"], axis=1, inplace=True)
    data_prepared = pd.merge(data_prepared, df_commande_min, on=["customer_unique_id"], how="inner")
    
    
    #prix moyen d'un produit
    df_prix_produit_moy = df_prix.groupby(["customer_unique_id"], as_index=False).mean()
    df_prix_produit_moy.rename(columns={"price": "prix_produit_moy"}, inplace=True)
    df_prix_produit_moy["ecart_type_produit_moyen"] = df_prix_produit_moy["prix_produit_moy"].std()
    df_prix_produit_moy.drop(["freight_value"], axis=1, inplace=True)
    data_prepared = pd.merge(data_prepared, df_prix_produit_moy, on=["customer_unique_id"], how="inner")
    
    #prix max d'un produit
    df_produit_max = df_prix.groupby(["customer_unique_id"], as_index=False).max()
    df_produit_max.rename(columns={"price": "prix_produit_max"}, inplace=True)
    df_produit_max.drop(["freight_value"], axis=1, inplace=True)
    data_prepared = pd.merge(data_prepared, df_produit_max, on=["customer_unique_id"], how="inner")
    
    #prix min d'un produit
    df_produit_min = df_prix.groupby(["customer_unique_id"], as_index=False).min()
    df_produit_min.rename(columns={"price": "prix_produit_min"}, inplace=True)
    df_produit_min.drop(["freight_value"], axis=1, inplace=True)
    data_prepared = pd.merge(data_prepared, df_produit_min, on=["customer_unique_id"], how="inner")
    
    
    #pourcentage moy FDP par rapport prix facture
    df_pourcent_moy = df_prix.groupby(["customer_unique_id"], as_index=False).mean()
    df_pourcent_moy["taux_moyen_fdp"] =  round((df_pourcent_moy["freight_value"] / df_pourcent_moy["price"]) * 100,2)
    df_pourcent_moy["ecart_type_taux_moyen_fdp"] = df_pourcent_moy["taux_moyen_fdp"].std()
    df_pourcent_moy.drop(["price", "freight_value"], axis=1, inplace=True)
    data_prepared = pd.merge(data_prepared, df_pourcent_moy, on=["customer_unique_id"], how="inner")
    
    return data_prepared
```

### Délai livraison moyen

```python
def delai_livraison(data_prepared):
    
    #On récupère les données de notre table unique
    df_delai = data.loc[:,["customer_unique_id", "order_purchase_timestamp", "order_delivered_customer_date"]]
    
    #On transfomre les champs en date
    df_delai["order_purchase_timestamp"] = pd.to_datetime(df_delai["order_purchase_timestamp"])
    df_delai["order_delivered_customer_date"] = pd.to_datetime(df_delai["order_delivered_customer_date"])
    
    #On calcule le délai de livarison pour chaque ligne
    df_delai["delai_livraison_moyen"] = (df_delai["order_delivered_customer_date"] - df_delai["order_purchase_timestamp"]).dt.days
    
    #On calcule le délai livraison moyen et on le merge au dataframe de sortie
    df_delai_moy = df_delai.groupby("customer_unique_id", as_index=False).mean()
    data_prepared = pd.merge(data_prepared, df_delai_moy, on=["customer_unique_id"], how="inner")
    
    #On calcule le délai livraison moyen et on le merge au dataframe de sortie
    df_delai_min = df_delai.groupby("customer_unique_id", as_index=False).min()
    df_delai_min.rename(columns={"delai_livraison_moyen": "delai_livraison_min"}, inplace=True)
    df_delai_min.drop(["order_purchase_timestamp", "order_delivered_customer_date"], axis=1, inplace=True)
    data_prepared = pd.merge(data_prepared, df_delai_min, on=["customer_unique_id"], how="inner")
    
    #On calcule le délai livraison moyen et on le merge au dataframe de sortie
    df_delai_max = df_delai.groupby("customer_unique_id", as_index=False).max()
    df_delai_max.rename(columns={"delai_livraison_moyen": "delai_livraison_max"}, inplace=True)
    df_delai_max.drop(["order_purchase_timestamp", "order_delivered_customer_date"], axis=1, inplace=True)
    data_prepared = pd.merge(data_prepared, df_delai_max, on=["customer_unique_id"], how="inner")
    
    return data_prepared
```

### Commentaires

```python
def commentaires(data_prepared):
    
    #On récupère les données de notre table unique
    df_comment = data.loc[:,["customer_unique_id", "review_comment_message"]]
    
    #On calcule la moyenne pour chaque customer_unique_id
    df_comment = df_comment.groupby(["customer_unique_id"], as_index=False).mean()
    
    #On transforme la variable de paiements étalés en booléen
    #Pour cela, on arrondie la moyenne calculée précédemment
    df_comment["review_comment_message"] = df_comment["review_comment_message"].apply(lambda x: 1 if x >= 0.5 else 0)
    df_comment["review_comment_message"].astype("bool")
    
    #On joint les 2 dataframes
    data_prepared = pd.merge(data_prepared, df_comment, on=["customer_unique_id"], how="inner")
        
    return data_prepared
```

### Features one hot

Nous allons tranformer les features onehot créées précédemment en features indiquant le taux de produits achetés par catégorie par rapport au nombre de commandes.

```python
def onehotfeatures(data_prepared):
    
    lst_col = [x for x in data.columns[data.columns.str.contains("customer_unique_id")]]
    lst_col1 = [x for x in data.columns[data.columns.str.contains("product_category_name")]]
    lst_col2 = [x for x in data.columns[data.columns.str.contains("payment_type")]]
    
    lst_col = lst_col + lst_col1 + lst_col2
    lst_col.remove("product_category_name_english")
    lst_col.remove("payment_type")
    
    print(lst_col[1:])
    #On récupère les données de notre table unique
    df_onehot = data.loc[:, lst_col]
    
    #On calcule la moyenne pour chaque customer_unique_id
    df_onehot = df_onehot.groupby(["customer_unique_id"], as_index=False).mean()
    
    #Si la moyenne du groupby est différent de 0, cela veut dire que la valeur était présente au moins une fois.
    #On passe donc la valeur à 1
    #Car cela veut dire que le client a acheté le produit au moins une fois
    #Ou utilisé ce moyen de paiement au moins une fois
    for col in df_onehot[lst_col[1:]]: 
        
        df_onehot[col] = df_onehot[col].apply(lambda x: 1 if x != 0 else 0)
        
        
    df_onehot[lst_col[1:]].astype("bool")
    
    data_prepared = pd.merge(data_prepared, df_onehot, on=["customer_unique_id"], how="inner")
    
    return data_prepared
```

### Taux catégories

Cette fonction permet de claculer le taux de produits acheté par clients par catégories par rapport à son nombre de commandes.

Idem pour les types de paiement.

```python
def taux_categories(data_prepared):
    
    lst_col = [x for x in data.columns[data.columns.str.contains("customer_unique_id")]]
    lst_col1 = [x for x in data.columns[data.columns.str.contains("taux_paiement")]]
    lst_col2 = [x for x in data.columns[data.columns.str.contains("taux_produit")]]
    
    lst_col = lst_col + lst_col1 + lst_col2
    
    #On récupère les données de notre table unique
    df_taux = data.loc[:, lst_col]
    
    #On additionne les lignes pour chaque customer_unique_id
    df_taux = df_taux.groupby(["customer_unique_id"], as_index=False).sum()
    
    data_prepared = pd.merge(data_prepared, df_taux, on=["customer_unique_id"], how="inner")
    
    #Gestion du total de chaque catégorie par client
    lst_col = ["customer_unique_id", "nbr_paiements_effectues", "nbr_produits_achetes"]
    
    df_nbr_categ = data.loc[:, lst_col]
    
    #On additionne les lignes pour chaque customer_unique_id
    df_nbr_categ = df_nbr_categ.groupby(["customer_unique_id"], as_index=False).sum()
    
    data_prepared = pd.merge(data_prepared, df_nbr_categ, on=["customer_unique_id"], how="inner")
    
    for x in lst_col1:
        data_prepared[x] = data_prepared[x].divide(data_prepared["nbr_paiements_effectues"], fill_value=1)
        
    for x in lst_col2:
        data_prepared[x] = data_prepared[x].divide(data_prepared["nbr_produits_achetes"], fill_value=1)

    return data_prepared
```

### Bucketing

Le bucketing est une méthode de classement des individus pour les identidier dans un histogramme.

Ici nous allons prendre la distribution de la variable "total dépenses" et nous allons créer 6 features :
 - moins de 30% de la distribution
 - entre 30% et 40% de la distribution
 - 40% - 50%
 - 50% - 60%
 - 60% - 70%
 - plus de 70%

```python
def bucketing2(data_prepared):
    
    df_bucketing = data_prepared.loc[:,["customer_unique_id","depenses_totales"]].copy()
    
    quantile_30 = df_bucketing["depenses_totales"].quantile(0.3)
    quantile_40 = df_bucketing["depenses_totales"].quantile(0.4)
    quantile_50 = df_bucketing["depenses_totales"].quantile(0.5)
    quantile_60 = df_bucketing["depenses_totales"].quantile(0.6)
    quantile_70 = df_bucketing["depenses_totales"].quantile(0.7)
    
    print(quantile_30)
    print(quantile_40)
    print(quantile_50)
    print(quantile_60)
    print(quantile_70)
    
    df_bucketing.reset_index()
    
    df_bucketing["moins_30%"] = np.where(df_bucketing["depenses_totales"] < quantile_30, 1, 0)
    
    df_bucketing["30%_40%"] = np.where(
        (df_bucketing["depenses_totales"] >= quantile_30) & 
        (df_bucketing["depenses_totales"] < quantile_40), 1, 0)
    
    df_bucketing["40%_50%"] = np.where(
        (df_bucketing["depenses_totales"] >= quantile_40) & 
        (df_bucketing["depenses_totales"] < quantile_50), 1, 0)
    
    df_bucketing["50%_60%"] = np.where(
        (df_bucketing["depenses_totales"] >= quantile_50) & 
        (df_bucketing["depenses_totales"] < quantile_60), 1, 0)
    
    df_bucketing["60%_70%"] = np.where(
        (df_bucketing["depenses_totales"] >= quantile_60) & 
        (df_bucketing["depenses_totales"] <= quantile_70), 1, 0)
    
    df_bucketing["plus_70%"] = np.where(df_bucketing["depenses_totales"] > quantile_70, 1, 0)
    
    df_bucketing.drop("depenses_totales", axis=1, inplace=True)
    
    data_prepared = pd.merge(data_prepared, df_bucketing, on=["customer_unique_id"], how="inner")
    
    return data_prepared
```

```python
def analyse_categ(df, col):

    mydata = df.loc[:, ["order_purchase_timestamp", col]]

    mydata["order_purchase_timestamp"] = pd.to_datetime(mydata["order_purchase_timestamp"])
    mydata["order_purchase_timestamp"] = mydata["order_purchase_timestamp"].dt.strftime("%d/%m/%Y").astype("datetime64")
    #df_analyse_fluctuations_achats = pd.to_timedelta(df_analyse_fluctuations_achats)

    mydata = mydata.loc[mydata[col] == 1]

    return mydata
    
```
